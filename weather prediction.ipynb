```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Load the historical weather data
data = pd.read_csv('JABALPUR.csv')
data['date_time']=pd.to_datetime(data['date_time'])
datetimedata=data['date_time']

# Select the relevant features for forecasting (temperature, humidity, pressure, etc.)
selected_features = ['WindGustKmph', 'cloudcover', 'humidity', 'precipMM', 'pressure','tempC','visibility','winddirDegree','windspeedKmph']
data = data[selected_features]

# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.values)
```

```python
# Define the number of previous time steps to consider
lookback = 30

# Prepare the data for training
X_train, y_train = [], []
for i in range(len(scaled_data) - lookback - 24):
    X_train.append(scaled_data[i:i+lookback])
    y_train.append(scaled_data[i+lookback:i+lookback+24])
X_train, y_train = np.array(X_train), np.array(y_train)

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(lookback, len(selected_features))))
model.add(LSTM(units=50))
model.add(Dense(units=24 * len(selected_features)))  # Output layer for the next 24 hours
model.compile(optimizer='adam', loss='mean_squared_error')
y_train = y_train.reshape(-1, 24 * len(selected_features))
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)

# Get the last lookback time steps from the training data
last_sequence = scaled_data[-lookback:]

# Prepare the input for forecasting
X_forecast = []
X_forecast.append(last_sequence)
X_forecast = np.array(X_forecast)

# Make the forecast
forecast = model.predict(X_forecast)

# Reshape the forecasted data
forecast = forecast.reshape(24, len(selected_features))

# Inverse scaling for the forecasted data
forecast = scaler.inverse_transform(forecast)

# Create a time index for the forecasted hours

```
